# Cleansing

```python
feat = '작업일자'
df1.drop(columns = [feat], axis = 0, inplace = True)
print(len(df1))
df1.head()
```

```python
df1 = df1.groupby(['사용월', '호선명', '지하철역']).agg({
    '유임승차인원': 'max',  # 유임승차인원
    '무임승차인원': 'max',  # 무임승차인원
    '유임하차인원': 'max',  # 유임하차인원
    '무임하차인원': 'max'  # 무임하차인원
}).reset_index().sort_values(by=['호선명', '지하철역'], ascending=[True, True])
df1.reset_index(drop=True, inplace=True)
print(len(df1))
df1.head()
```

```python
# 날짜형 변수의 경우, 'year'과 'month' feature로 나눠서 저장.
feat = '사용월'
df1[feat] = pd.to_datetime(df1[feat], format='%Y%m')
df1['year'] = df1[feat].dt.year.astype(int)
df1['month'] = df1[feat].dt.month.astype(int)

df1.drop(columns = [feat], axis = 0, inplace = True)
print(df1.head())
```

```python
station_list = sorted(df1['지하철역'].unique().tolist())
print(station_list)

df1['지하철역'] = df1['지하철역'].apply(lambda x: re.sub(r'\(.*\)', '', x))

station_list = sorted(df1['지하철역'].unique().tolist())
print(station_list)
```

# 역 정보

```python
csv = dir_name+"data_loc.csv"     # train data set to csv

df2 = pd.read_csv(csv)                          # data frame of csv, i.e. source file

df2.rename(columns={'lineNm': '호선명', 'stnKrNm': '지하철역'}, inplace=True)
df2.head()
```

```python
df = pd.merge(df1, df2[['지하철역', '호선명', 'convX', 'convY']], on=['지하철역', '호선명'], how='left')
df.head()
```

```python
df.sort_values(by=['year','month', '지하철역', '호선명'], ascending=[False, False, True, True], inplace=True)
print(len(df))
df.head()
```

```python
df.groupby(['year', 'month', '지하철역']).agg({
    '호선명': 'first',  # 호선명은 그룹의 첫 번째 값을 사용
    '유임승차인원': 'sum',  # 유임승차인원 합계
    '무임승차인원': 'sum',  # 무임승차인원 합계
    '유임하차인원': 'sum',  # 유임하차인원 합계
    '무임하차인원': 'sum',  # 무임하차인원 합계
    'convX': 'mean',
    'convY': 'mean'
}, inplace=True).reset_index().sort_values(by=['year','month', '호선명', '지하철역'], ascending=[False, False, True, True])
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
line_list = sorted(df1['호선명'].unique().tolist())
print(line_list)

station_list = sorted(df1['지하철역'].unique().tolist())
print(station_list)

year_list = sorted(df1['year'].unique().tolist())
print(year_list)

month_list = list(range(1,13))
print(month_list)

avail_date = [[0,0],[0,0]]

for l in line_list :
  station_list = sorted(df1.loc[df1['호선명'] == l, '지하철역'].unique().tolist())

  for s in station_list :
    flag = False

    for y in year_list :
      for m in month_list :
        # if bool(len(df1[(df1['호선명'] == l) & (df1['지하철역'] == s) & (df1['year'] == y) & (df1['month'] == m)])) == True :
        if len(df1[(df1['호선명'] == l) & (df1['지하철역'] == s) & (df1['year'] == y) & (df1['month'] == m)]) == 1 :
          if not flag :
            avail_date[0] = [y,m]
            flag = True
          year_before = y
          month_before = m
        elif len(df1[(df1['호선명'] == l) & (df1['지하철역'] == s) & (df1['year'] == y) & (df1['month'] == m)]) > 1 :
          print(f"{l}, {s}\t{df1[(df1['호선명'] == l) & (df1['지하철역'] == s) & (df1['year'] == y) & (df1['month'] == m)]}")
        else :
          if flag :
            avail_date[1] = [year_before, month_before]
            flag = False
    if flag :
      avail_date[1] = [y, m]

    print(f"{l}, {s} : {avail_date[0][0]}.{avail_date[0][1]} ~ {avail_date[1][0]}.{avail_date[1][1]}")
```
